```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

# 假设的数据 (64x8)
data = np.random.rand(64, 8)

# 创建图
G = nx.Graph()

# 添加上面的节点
for i in range(64):
    G.add_node(f"upper_{i}")

# 添加下面的节点
for i in range(8):
    G.add_node(f"lower_{i}")

# 添加边
for i in range(64):
    for j in range(8):
        if data[i, j] > 0.5:
            G.add_edge(f"upper_{i}", f"lower_{j}")

# 绘制图
pos = {}
for i in range(64):
    pos[f"upper_{i}"] = (i, 1)
for i in range(8):
    pos[f"lower_{i}"] = (i * 8, 0)

plt.figure(figsize=(12, 6))
nx.draw(G, pos, with_labels=False, node_size=20, node_color="blue", font_size=12, font_color="black", edge_color="gray")

# 在每个节点上绘制曲线
for i in range(64):
    x = np.linspace(i, i+1, 8)
    y = data[i, :]
    plt.plot(x, y + 1, color='red')

for i in range(8):
    x = np.linspace(i * 8, (i * 8) + 1, 64)
    y = data[:, i]
    plt.plot(x, y, color='green')

plt.show()
```
